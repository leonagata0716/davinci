```python
alt.Chart(dfmap).mark_bar().encode(
    x = 'Police',
    y = 'sum(Crimes)'
)
```

```python
alt.Chart(dfmap).mark_bar().encode(
    x = 'sum(Crimes)',
    y = 'Police',
)
```

```python
alt.Chart(dfmap).mark_bar(color = '#FF5733').encode(
    alt.X('Crimes', 
          type = 'quantitative', 
          aggregate = 'sum',
          title = 'Crimes'),
    alt.Y('Police', 
          type = 'nominal',
          title = 'Police Station')
)
```

```python
alt.Chart(dfmap).mark_bar(color = '#FF5733').encode(
    alt.X('Crimes:Q', 
          aggregate = 'sum',
          title = 'Crimes'),
    alt.Y('Police:N', 
          title = 'Police Station')
)
```

```python
alt.Chart(dfmap).mark_bar(color = '#FF5733').encode(
    alt.X('Crimes:Q', 
           bin = True,
           title = 'Crimes'),
    alt.Y('Police:N', 
           aggregate = 'count',
           title = 'Count of Records')
)
```

```python
counties = alt.topo_feature(data.us_10m.url, feature='counties')
```

```python
# US states background
background = alt.Chart(counties).mark_geoshape(
    fill='#FBEAE6',
    stroke='white'
).properties(
    width=700,
    height=500
).project('albersUsa')
```

```python
# plots positions on background
points = alt.Chart(dfmap).transform_aggregate(
    latitude='mean(latitude)',
    longitude='mean(longitude)',
    count='sum(Crimes)'
).mark_circle().encode(
    longitude='longitude:Q',
    latitude='latitude:Q',
    size=alt.Size('count:Q', title='Count of Crimes'),
    color=alt.value('#009CFF'),
    tooltip=['count:Q']
).properties(
    title='Count of Crimes'
)

background + points
```
